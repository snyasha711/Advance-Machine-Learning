```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

## for feature slection

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

# to visualise al the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
dataset=pd.read_csv('traindone.csv')
```

```python
dataset.head()
```

```python
ytraindonedep=dataset[['SalePrice']]
```

```python
traindone=dataset.drop(['Id','SalePrice'],axis=1)
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) # remember to set the seed, the random state in this function
feature_sel_model.fit(traindone, ytraindonedep)
```

```python
feature_sel_model.get_support()
```

```python
selected_feat = traindone.columns[(feature_sel_model.get_support())]

# let's print some stats
print('total features: {}'.format((traindone.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrank to zero: {}'.format(
    np.sum(feature_sel_model.estimator_.coef_ == 0)))
```

```python


selected_feat

```

```python
traindone=traindone[selected_feat]
```

```python
traindone.head()
```

```python
ytraindonedep
```

```python
from sklearn.model_selection import GridSearchCV,train_test_split

X_train, X_test, y_train, y_test = train_test_split(traindone,ytraindonedep,
                                                    test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression  
regressor= LinearRegression()  
regressor.fit(X_train, y_train)  
```

```python
print("B1=",regressor.coef_)
print("B0 =",regressor.intercept_)
```

```python
y_pred = regressor.predict(X_test)
```

```python
print(y_test)
print()
print(y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred1 = regressor.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred1)
print("Mean squared error:", mse)
```

```python
r2_score(y_test, y_pred1)
```

```python
parameters = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}

from sklearn.linear_model import Ridge, Lasso
ridge = Ridge()
lasso = Lasso()
```

```python
ridge_cv = GridSearchCV(ridge, parameters, cv=5)
lasso_cv = GridSearchCV(lasso, parameters, cv=5)

ridge_cv.fit(X_train, y_train)
lasso_cv.fit(X_train, y_train)
```

```python
ridge_score = ridge_cv.score(X_test, y_test)
lasso_score = lasso_cv.score(X_test, y_test)
```

```python
print('Ridge Best Alpha:', ridge_cv.best_params_)
print('Ridge Score:', ridge_score)
print('Lasso Best Alpha:', lasso_cv.best_params_)
print('Lasso Score:', lasso_score)
```

```python
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred_ridge)
```

```python
lasso = Lasso(alpha=0.001)
lasso.fit(X_train, y_train)
y_pred_lasso = lasso.predict(X_test)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred_lasso)
```

```python
print("Ridge MSE: {:.4f}".format(mse_ridge))
print("Lasso MSE: {:.4f}".format(mse_lasso))
```

Ridge is better
